```python
import cv2  #For image reading
```

```python
#tqdm shows the estimated time remaining for the iterable.
# def is the keyword for defining a function.
#cv2. imread() method loads an image from the specified file.
#cv2.cvtColor() method is used to convert an image from one color space to another
#cv2.resize() is helps in reducing the number of pixels from an image 
#cv2.COLOR_BGR2RGB converting bgr to rgb 

def get_images(data,batch_size):
  images=[]
  for i in tqdm(range(0,len(data),batch_size)):
    # print(i)
    for k in data[i:i+batch_size]:
        #print(k)
        img=cv2.imread(k)
        img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        res=cv2.resize(img,(256,256),interpolation=cv2.INTER_AREA)
        images.append(res)
  return images
```

```python
#With the help of np.expand_dims() method, we can get the expanded dimensions 
#preproces_input  meant to adequate your image to the format the model requires. Some models use images with values ranging from 0 to 1
# flatten is converting multipledim to onedim
#model.pridect is use to predict the labels of the data values on the basis of the trained model.
from tqdm import tqdm
def get_deep_features(data,model):
    feats=[]
    for i in tqdm(data):
        #print(i)
        x_in = np.expand_dims(i, axis=0)
        x_in = preprocess_input(x_in)
        features = model.predict(x_in)
        features = features.flatten()
        feats.append(features)
    return feats
```
